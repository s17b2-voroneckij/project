```python
from scipy import integrate,spatial
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f1(t,X,rho = 28, sigma = 10, beta = 2.66):
    x,y,z = X
    dx = sigma*y- sigma*x
    dy = rho*x - y - x*z
    dz = x*y - beta * z
    return dx,dy,dz
```

```python
system = integrate.solve_ivp(f1,(0,1000),t_eval=np.linspace(0,1000,5000),y0=[0,0.1,1.05],method='RK45')
```

```python
x_r,y_r,z_r = system.y[0], system.y[1],system.y[2]
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(x_r, y_r, z_r, lw=0.5)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Lorenz Attractor")
```

```python
plt.figure(figsize=(20,7))
plt.title("Lorenz Series, rho = 28")
plt.plot(np.linspace(0,1000,5000),x_r)
```
