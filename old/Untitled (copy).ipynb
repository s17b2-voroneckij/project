```python
def draw_quality(k, hs, filename):
    quality = []
    q1 = []
    q2 = []
    q3 = []
    q4 = []
    for h in hs:
        set_params(k, h)
        f = open(filename)
        readpoints(f)
        clusters, cluster, g = Wishart()
        quality.append(quality_cheat(cluster))
        qo = quality_honest(clusters, cluster)
        q1.append(qo[0])
        q2.append(qo[1])
        q3.append(qo[2])
        q4.append(qo[3])
    plt.plot(hs, quality, color='black')
    plt.plot(hs, q1, color='yellow')
    plt.plot(hs, q3, color='pink')
    plt.plot(hs, q2, color='orange')
    plt.plot(hs, q4, color='red')
```

```python
hs = np.arange(0, 3, 0.2)
```

```python
draw_quality(1, hs, 'data_15_dim_2.txt')
```

```python
k = 3
nc = []
for h in hs:
    set_params(k, h)
    f = open('data_15_dim_2.txt')
    readpoints(f)
    clusters, cluster, g = Wishart()
    curr = 0
    for cl in clusters:
        if not cl.is_deleted:
            curr += 1
    nc.append(curr)
plt.plot(hs, nc)
```
