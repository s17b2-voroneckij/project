```python
def num_clusters(LSADict, filenames, z=3, dim=5, k=5, h=0.5):
    clusters, cluster, g = make_paths(LSADict, filenames, z, dim, k, h)
    return len(set(cluster)), cluster.count(0) / len(cluster)
```

```python
def draw(LSADict, filenames, z, k, h, dim):
    nc = []
    trash = []
    for d in dim:
        num, tr = num_clusters(LSADict, filenames, z, d, k, h)
        nc.append(num)
        trash.append(tr)
    plt.plot(dim, nc)
    return trash
```
